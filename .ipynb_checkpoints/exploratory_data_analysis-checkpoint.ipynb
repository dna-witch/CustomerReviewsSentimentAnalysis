## Text Preprocessing

The text preprocessing steps include:
1. Lowercasing the text
2. Removal of HTML tags
3. Removal of special characters
4. Expansion of contractions
5. Removal of stopwords
6. Tokenization
7. Lemmatization
8. Create a vocabulary

```python
def lowercase(text):
    '''This function converts text to lowercase.'''
    return text.str.lower()
```

```python
def remove_special_characters(text):
    '''This function removes special characters from a string.'''
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text
```

```python
def remove_html_tags(text):
    '''This function removes HTML tags from a string.'''
    soup = BeautifulSoup(text, 'html.parser')
    return soup.get_text()
```

```python
def remove_stopwords(text):
    '''This function removes stopwords from a string.'''
    stop_words = set(nltk.corpus.stopwords.words('english'))
    words = word_tokenize(text)
    words = [word for word in words if word not in stop_words]
    text = ' '.join(words)
    return text
```

```python
def expand_contractions(text):
    '''This function expands contractions in a string.'''
    expanded_text = contractions.fix(text)
    return expanded_text
```

```python
def tokenize(text):
    '''This function tokenizes a string.'''
    tokenizer = nltk.tokenize.WhitespaceTokenizer()
    tokens = tokenizer.tokenize(text)
    return tokens
```

```python
def lemmatize(tokens):
    '''This function lemmatizes words.'''
    lemmatizer = nltk.stem.WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return lemmatized_tokens
```

```python
def create_vocabulary(tokens):
    '''This function creates a vocabulary from a list of tokens.'''
    vocabulary = set([word for token in tokens for word in token])
    return vocabulary
```

```python
def preprocess(series):
    # Lowercasing the text
    series = lowercase(series)
    # Expansion of contractions
    series = series.apply(expand_contractions)
    # Removal of special characters
    series = series.apply(remove_special_characters)
    # Removal of stopwords
    series = series.apply(remove_stopwords)
    # Store the preprocessed text data in a new attribute
    preprocessed_text = series
    print('Text data preprocessed successfully!')

    # Lemmatization: transforms the words into their semantically correct base forms
    tokens = series.apply(lemmatize)
    # Tokenization: splits the text into individual words
    tokens = series.apply(tokenize)
    print('Text data tokenized and lemmatized successfully!')

    # # Create a vocabulary
    # vocabulary = create_vocabulary(tokens)
    # print('Vocabulary created successfully!')
    return tokens
```

```python
# Remove HTML tags
df['text'] = df['text'].apply(remove_html_tags)
```

### Verified vs Unverified Purchases

Since technically anyone can review any product on the Amazon website, a "verified purchase" tag signals to readers that this user's review is trustworthy because Amazon has confirmed that they actually purchased and received the product. This is important to counteract bias from sellers who might list their products and try to skew their product ratings by buying positive reviews.

```python
verification_counts = df['verified_purchase'].value_counts()

# Plotting the distribution of verified_purchase
plt.figure(figsize=(10, 5))
verification_counts.plot(kind='bar', color=['lightblue', 'coral'])
plt.title('Verified vs Unverified Purchases')
plt.xlabel('Verification Status')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

#### How many verified and unverified purchases do the top reviewers have?

This will give us an estimated metric of how trustworthy these reviewers are. This is not to say that reviews from unverified purchases are all inherently false or untrustworthy, but that is a risk that the reader will have to take.

```python
review_counts = df['user_id'].value_counts()
frequent_reviewers = review_counts[review_counts > 1000].index

filtered_df = df[df['user_id'].isin(frequent_reviewers)]
grouped_df = filtered_df.groupby(['user_id', 'verified_purchase']).size().unstack(fill_value=0)

plt.figure(figsize=(12, 4))
grouped_df.plot(kind='bar', stacked=True)
plt.title('Verified vs Unverified Purchases of Frequent Reviewers (Over 1000 Reviews)')
plt.xlabel('User ID')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Verified Purchase')
plt.show()
```

Out of these 8 frequent reviewers, 7 of them have significantly more unverified purchases than verified purchases associated with their reviews. Only 2 of these frequent reviewers have more verified purchases than unverified. 

### Distribution of Ratings

```python
ratings_counts = df['rating'].value_counts().sort_index()

plt.figure(figsize=(10, 8))
ratings_counts.plot(kind='bar', color='purple')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding count values on top of each bar
for i, count in enumerate(ratings_counts):
    plt.text(i, count + 10, str(count), ha='center', va='bottom')

plt.show()
```

```python
rating_percentages = df['rating'].value_counts(normalize=True).sort_index() * 100

for i, percentage in enumerate(rating_percentages):
    print(f'Out of all ratings, {i+1}-star are {percentage:.2f} of the total.')
```

There are significantly more 5-star ratings than any other rating. This skew towards positive reviews might bias our sentiment analysis model, since there are not as many "negative" reviews to learn from. To counter this potential bias, we can employ data stratification when creating our training dataset.

#### Verified Purchases per Rating

```python
percentage_verified_purchase = df.groupby('rating')['verified_purchase'].value_counts(normalize=True) * 100

# Selecting the percentage of verified purchases for 'True' (verified) purchases
percentage_verified_true = percentage_verified_purchase[:, True]

print("Percentage of Verified Purchases for Each Rating:")
print(percentage_verified_true)
```

There isn't too much difference between the proportions of verified amongst the rating brackets (which is good), so purchase verification might not actually give us much more useful information. 

It is also good to note that the verified purchase percentage is over 50% for all rating brackets!

## Design Decisions

To account for the imbalanced distribution of reviews across rating brackets, one solution could be to employ data stratification when creating the training dataset. Otherwise, we risk bias in the model since ~63% of the reviews are positive (3-stars and up).
